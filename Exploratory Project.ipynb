```python
# Imports
# 
# Note: these are all the imports you need for this assignment!
# Do not import any other functions / packages

# Display plots directly in the notebook instead of in a new window
%matplotlib inline

# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configure libraries
# The seaborn library makes plots look nicer
sns.set()
sns.set_context('talk')

# Don't display too many rows/cols of DataFrames
pd.options.display.max_rows = 7
pd.options.display.max_columns = 8

# Round decimals when displaying DataFrames
pd.set_option('precision', 2)
```

```python
# Read all csv
stats_game_2000 = pd.read_csv("stats/00-01_team_per_game_stats.csv")
stats_shooting_2000 = pd.read_csv("stats/00-01_team_shooting.csv")

# All team stats for all years
df_stats_team_stats_2000 = pd.read_csv("stats/00-01_team_stats.csv")
df_stats_team_stats_2001 = pd.read_csv("stats/01-02_team_stats.csv")
df_stats_team_stats_2002 = pd.read_csv("stats/02-03_team_stats.csv")
df_stats_team_stats_2003 = pd.read_csv("stats/03-04_team_stats.csv")
df_stats_team_stats_2004 = pd.read_csv("stats/04-05_team_stats.csv")
df_stats_team_stats_2005 = pd.read_csv("stats/05-06_team_stats.csv")
df_stats_team_stats_2006 = pd.read_csv("stats/06-07_team_stats.csv")
df_stats_team_stats_2007 = pd.read_csv("stats/07-08_team_stats.csv")
df_stats_team_stats_2008 = pd.read_csv("stats/08-09_team_stats.csv")
df_stats_team_stats_2009 = pd.read_csv("stats/09-10_team_stats.csv")
df_stats_team_stats_2010 = pd.read_csv("stats/10-11_team_stats.csv")
df_stats_team_stats_2011 = pd.read_csv("stats/11-12_team_stats.csv")
df_stats_team_stats_2012 = pd.read_csv("stats/12-13_team_stats.csv")
df_stats_team_stats_2013 = pd.read_csv("stats/13-14_team_stats.csv")
df_stats_team_stats_2014 = pd.read_csv("stats/14-15_team_stats.csv")
df_stats_team_stats_2015 = pd.read_csv("stats/15-16_team_stats.csv")
df_stats_team_stats_2016 = pd.read_csv("stats/16-17_team_stats.csv")
df_stats_team_stats_2017 = pd.read_csv("stats/17-18_team_stats.csv")
df_stats_team_stats_2018 = pd.read_csv("stats/18-19_team_stats.csv")

# Create new dataframe with all team stats with 3PA
df_all_team_stats = pd.DataFrame()

years = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008','2009', '2010', '2011', '2012', 
         '2013', '2014', '2015', '2016', '2017', '2018'] 
df_all_team_stats['Year'] = years

# Get sum of all 3PA for team each for all years
three_attempts = [df_stats_team_stats_2000['3PA'].sum(), df_stats_team_stats_2001['3PA'].sum(), 
                  df_stats_team_stats_2002['3PA'].sum(), df_stats_team_stats_2003['3PA'].sum(),
                  df_stats_team_stats_2004['3PA'].sum(), df_stats_team_stats_2005['3PA'].sum(),
                  df_stats_team_stats_2006['3PA'].sum(), df_stats_team_stats_2007['3PA'].sum(),
                  df_stats_team_stats_2008['3PA'].sum(), df_stats_team_stats_2009['3PA'].sum(), 
                  df_stats_team_stats_2010['3PA'].sum(), df_stats_team_stats_2011['3PA'].sum(), 
                  df_stats_team_stats_2012['3PA'].sum(),
                  df_stats_team_stats_2013['3PA'].sum(), df_stats_team_stats_2014['3PA'].sum(),
                  df_stats_team_stats_2015['3PA'].sum(), df_stats_team_stats_2016['3PA'].sum(),
                  df_stats_team_stats_2017['3PA'].sum(), df_stats_team_stats_2018['3PA'].sum()]
df_all_team_stats['3PA'] = three_attempts

# Get sum of all 2PA for team each for all years
two_attempts = [df_stats_team_stats_2000['2PA'].sum(), df_stats_team_stats_2001['2PA'].sum(), 
                  df_stats_team_stats_2002['2PA'].sum(), df_stats_team_stats_2003['2PA'].sum(),
                  df_stats_team_stats_2004['2PA'].sum(), df_stats_team_stats_2005['2PA'].sum(),
                  df_stats_team_stats_2006['2PA'].sum(), df_stats_team_stats_2007['2PA'].sum(),
                  df_stats_team_stats_2008['2PA'].sum(), df_stats_team_stats_2009['2PA'].sum(), 
                  df_stats_team_stats_2010['2PA'].sum(), df_stats_team_stats_2011['2PA'].sum(), 
                  df_stats_team_stats_2012['2PA'].sum(),
                  df_stats_team_stats_2013['2PA'].sum(), df_stats_team_stats_2014['2PA'].sum(),
                  df_stats_team_stats_2015['2PA'].sum(), df_stats_team_stats_2016['2PA'].sum(),
                  df_stats_team_stats_2017['2PA'].sum(), df_stats_team_stats_2018['2PA'].sum()]
df_all_team_stats['2PA'] = two_attempts



#print(df_all_team_stats)
```

```python
# Graphs

# Univariate --> histogram of one variable (ex: 3-point team attempts) (then I need a graph for each year, so 19 graphs?)

# Bivariate --> boxplot or scatterplot (ex: 3-point team shooting and 2-point team shooting attempts)

# Shows all 3PA for all teams in each year
df_all_team_stats.plot(x='Year', y='3PA', xticks=range(19), title="3PA Over Past Two Decades")

f4 = plt.gcf()
plt.gca().margins(x=0)
f4.canvas.draw()
tl = plt.gca().get_xticklabels()
maxsize = max([t.get_window_extent().width for t in tl])
m = 0.2 # inch margin
s = maxsize/plt.gcf().dpi*19+2*m
margin = m/plt.gcf().get_size_inches()[0]

plt.gcf().subplots_adjust(left=margin, right=1.-margin)
plt.gcf().set_size_inches(s, plt.gcf().get_size_inches()[1])

# Dimensionality Reduction --> ICA or PCA or clustering

```
